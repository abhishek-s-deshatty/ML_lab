```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score
df=load_breast_cancer()
```

```python
x=df.data
y=df.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
ntrees=[1,5,10,20,50,100,200]
accu_scores=[]
prec_scores=[]
f1_scores=[]
recall_scores=[]
for n in ntrees:
    rfc=RandomForestClassifier(n_estimators=n,random_state=42)
    rfc.fit(x_train,y_train)
 
    y_pred=rfc.predict(x_test)
    
    acc=accuracy_score(y_test, y_pred)
    pr=precision_score(y_test, y_pred,average='weighted')
    rec=recall_score(y_test, y_pred,average='weighted')
    f1=f1_score(y_test, y_pred,average='weighted')
 
    accu_scores.append(acc)
    prec_scores.append(pr)
    recall_scores.append(rec)
    f1_scores.append(f1)
 
    print(f'No of trees: {n}')
    print(f'Accuracy: {acc:.2f}')
    print(f'Precision: {pr:.2f}')
    print(f'Recall: {rec:.2f}')
    print(f'F1_score: {f1:.2f}\n')
```

```python
plt.figure(figsize=(6,6))
plt.plot(ntrees,accu_scores,label='Accuracy',marker='o')
plt.plot(ntrees,prec_scores,label='Precision',marker='.')
plt.plot(ntrees,recall_scores,label='Recall',marker='d')
plt.plot(ntrees,f1_scores,label='F1',marker='s')
plt.xlabel("No of trees")
plt.ylabel("Score")
plt.title("Random forest classifier Breast Cancer Dataset")
plt.legend()
plt.grid(True)
plt.show()
```
